```python
import pandas as pd
```

```python
df_kg = pd.read_excel("D:/guided_repair/Guided-Repair/Knowledge_graph_dtcs.xlsx")
```

```python
df_kg = df_kg[['grouped_dtc', 'clean_verified_subsystem']]
```

```python
df_kg = df_kg[(df_kg['clean_verified_subsystem']!='not considering')]
```

```python
df_kg.shape[0]
```

```python
import ast
df_dtc = pd.DataFrame({'dtc':[], 'subsystem':[]})
dtc_ls = []
count = 0

for i in range(df_kg.shape[0]):
    print(i)
    grouped_dtc_ls = ast.literal_eval(df_kg['grouped_dtc'].iloc[i])
    dtc_ls.extend(grouped_dtc_ls)
    for dtc in grouped_dtc_ls:
        df_dtc.loc[count, 'dtc'] = dtc
        df_dtc.loc[count, 'subsystem'] = df_kg['clean_verified_subsystem'].iloc[i]
        count+=1
    print(df_dtc)
    print(i, " index complete")
```
